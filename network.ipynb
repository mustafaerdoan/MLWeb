```python
import networkx as nx
import matplotlib.pyplot as plt

# Parameters
n_nodes = 100
n_edges = 300

# Generate ErdsRnyi graph
er_graph = nx.gnm_random_graph(n_nodes, n_edges)

# Generate BarabsiAlbert graph
ba_graph = nx.barabasi_albert_graph(n_nodes, int(n_edges/n_nodes))

# Plot the graphs
plt.figure(figsize=(12, 6))

plt.subplot(121)
nx.draw(er_graph, node_size=30, with_labels=False)
plt.title("ErdsRnyi Graph")

plt.subplot(122)
nx.draw(ba_graph, node_size=30, with_labels=False)
plt.title("BarabsiAlbert Graph")

plt.show()
```

```python
# Compute PageRank
er_pagerank = nx.pagerank(er_graph)
ba_pagerank = nx.pagerank(ba_graph)

# Display top 5 nodes by PageRank for both models
er_top_nodes = sorted(er_pagerank.items(), key=lambda x: x[1], reverse=True)[:5]
ba_top_nodes = sorted(ba_pagerank.items(), key=lambda x: x[1], reverse=True)[:5]

print("Top PageRank nodes for ErdsRnyi Graph:", er_top_nodes)
print("Top PageRank nodes for BarabsiAlbert Graph:", ba_top_nodes)
```

```python
# Plot PageRank distributions
plt.figure(figsize=(12, 6))

plt.subplot(121)
plt.hist(list(er_pagerank.values()), bins=30, color='blue', alpha=0.7)
plt.title("PageRank Distribution for ErdsRnyi Graph")

plt.subplot(122)
plt.hist(list(ba_pagerank.values()), bins=30, color='green', alpha=0.7)
plt.title("PageRank Distribution for BarabsiAlbert Graph")


#Comparison will show that ErdsRnyi graphs have a more uniform PageRank distribution, while BarabsiAlbert graphs have a skewed distribution, where a few nodes dominate in terms of PageRank.
plt.show()
```
